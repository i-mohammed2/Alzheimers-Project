```python
import cv2
from ultralytics import YOLO
import pandas as pd

model = YOLO('yolov8m-pose.pt')

results = model(r"vid1_cam2_cropped.mp4")

```

```python
import numpy as np
import pandas as pd
dfs = []
#save the results.keypoints.data which is an array of [x, y, confidence] for each frame
for i, r in enumerate(results):
    keypoints = r.keypoints.numpy().data
    #flatten the array to a 2d array
    keypoints = np.array(keypoints).reshape(-1, 3)
    #convert the array to a pandas dataframe
    df = pd.DataFrame(keypoints, columns=['x', 'y', 'confidence'])
    #add a frame column
    df['frame'] = i
    #add a joint column: [nose, left_eye, right_eye, left_ear, right_ear, left_shoulder, right_shoulder, left_elbow, right_elbow, left_wrist, right_wrist, left_hip, right_hip, left_knee, right_knee, left_ankle, right_ankle]
    df['joint'] = np.tile(['nose', 'left_eye', 'right_eye', 'left_ear', 'right_ear', 'left_shoulder', 'right_shoulder', 'left_elbow', 'right_elbow', 'left_wrist', 'right_wrist', 'left_hip', 'right_hip', 'left_knee', 'right_knee', 'left_ankle', 'right_ankle'], len(df) // 17)
    #append the dataframe to the list
    dfs.append(df)

#concatenate the dataframes
df = pd.concat(dfs)
#save the dataframe to a csv file
df.to_csv('keypoints_cam2_vid1.csv', index=False)
```
